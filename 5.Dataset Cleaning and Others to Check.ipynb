## Duplicated and Constant Features

````python
row = [[True,13,'H',1.2,1.2,'A','C'], [True,13,'H',36.6,36.6,'B','A'],[False,13,'H',0,0,'A','C'], [False,13,'G',-14,-14,'C','B']]
traintest = pd.DataFrame(row, columns = ['is_train','f0','f1','f2','f3','f4','f5'])
traintest
# traintest.nunique(axis=1) == 1
````

**f0** : You can see f0 feature is constant for train and test data, it is not useful for the model !
<t> **f1** : f1 feature has constant value in train set, but not constant in test set. So it is not useful too for the model !
<t> if the categorical value is not constant in trainset, but there are value only in the test set, we need to handle this situation !
- Do this value matter much or not ? Can simulate this situation on validation set and compare the quality without and with the new feature values
- Or make the separate model for that object with the new feature values !
- Remove the feature values and see the quality of validation set !
    
<t> **f2,f3** : There are 2 duplicate numeric feature that are the same, it is just slow the model to train.
