## Art Tools

### Histograms
- Split feature range into bins and show how many points fall into each bin.
- Histogram maybe misleading in some cases. So,try to variate number of bins when using it.
- Note : Histogram aggregate the data, so we cannot see if there is a repeated value or unique values.

```python
import matplotlib.pyplot as plt
```

```python
data.columns
```

```python
plt.hist(data.price);
```

You can see there is zero value in this feature ! but actually it isn't

- Suggestions : Never make a conclusion based on a single plot !
- If you have hypothesis, try to make several different plots to prove it.

```python
plt.hist(data.price/ 100000, bins = 50);
```

What is the meaning of the peak of a histogram ?

Peak is the location of mean values of the feature
- If in a histogram there is a unique peak, that means the feature has a missing value that organization filled it with mean value.
- This information can be use for generate new feature like isnull() or using XgBoost Model !

### Plot (index versus value)

```python
data.trip.replace(['trip','roundtrip','round'], [1,2,3], inplace=True)
plt.scatter(range(len(data.price)), data.price, c= data.trip);
```

## Explore feature Relations
- It is hard to make conclusion in one feature, so see the correlation !

```python
data.columns
```

```python
plt.scatter(data.price, data.member_duration_days, c = data.trip);
```

```python
data['trip_'] = data.trip
test['trip_'] = 0
full = pd.concat([data,test], keys = ['train','test'])
```

```python
plt.scatter(full.price, full.member_duration_days, c = full.trip_);
```

- You can see the two scatter plot, before and after concatenate. From these plot we can know the distribution based on trip features for each test and train   set.

- This is important, if the distribution of train set and test set is not the same, we may overfit.
